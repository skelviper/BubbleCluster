read pairs file and output intra contacts by chromosomes

```r
file_names <- dir("./pairs/", 
pattern = "*.pairs", recursive = F, full.names = T)
```

```r
length(file_names)
```

```r
for (j in 1:125){
    df<-read_table2(file_names[j]) 
    for(i in list("chr1","chr2","chr3","chr4","chr5","chr6","chr7","chr8","chr9","chr10","chr11","chr12","chr13","chr14","chr15","chr16","chr17","chr18","chr19","chrX"))
    {
        write_tsv(df%>% filter(X1==X2 & X1 == i) %>% mutate(pos1 = floor(pos1/1000000),pos2 = floor(pos2/1000000)) %>% group_by(pos1,pos2) %>% summarise(count=n()),paste0("contactMatrix/",substr(file_names[j],10,99),"_",i,".txt"),col_names = F)
    }
    print(j)
}

```

```r
floor(999999/1000000)
```

```r
# 从第二个文件开始合并
for (i in 2:length(file_names)) {
    newdf <- read_table2(file_names[i],,col_names = F) %>% filter(X1 %in% chrList)
    newdf $ X1 <-as.character(newdf $ X1)
    names(newdf)<-c("chr","bin",substr(strsplit(file_names[i],split='.pairs')[[1]][1],11,99))
    df <- df %>% full_join(newdf,by=c("chr","bin"))
    }

names(df)[1:2] <- c("chr","bin")
write_tsv(df,"./binCPG.tsv")
```
